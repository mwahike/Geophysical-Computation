```python
print("FINDING ROOT OF x * sin(x) - e^x WHILE LOOP")
print()
#import library
import math

#define a function
def func(x): 
    return x * math.sin(x) - math.e**x

#function for calling bisection method
# a = upper bound
# b = lower bound
# c = center
def bisection(a, b, tol):
    i=0
    err=abs(b-a)
    
    #making decision
    if func(a)*func(b)>0: 
        print("The root cant be found, please change a and b")
        return None
    else:
        while err>tol:
            c=(a+b)/2
            i=i+1
            print("{:2d}  {:3.4f}  {:3.4f}  {:3.4f}  {:3.4f}  {:3.4f}  {:3.4f}  {:3.4f})"
                .format(i,a,b,c,func(a),func(b),func(c),err))
            
            if func(c)==0:
                return c
            elif func(c) < 0:
                b=c
            else:
                a=c
            err = abs(b-a)
        return c

print("iter   a    b    c    f(a)    f(b)    f(c)    err")
print("...................................................")
c= bisection(-1, 0, 1e-5)
if c is not None:
    print("...................................................")
    print(f"The root of f(x) is", round(c,6))
else:
    print("Root finding failed.")
```
